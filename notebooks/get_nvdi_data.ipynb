```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import glob

tif_files = glob.glob("../data/raw/modis-terra vegetation indices/nvdi/MOD13Q1.061__250m_16days_NDVI_2024*.tif")

with rasterio.open(tif_files[0]) as src:
    ndvi = src.read(1)  
    profile = src.profile

ndvi = ndvi.astype(float)
ndvi[ndvi < -2000] = np.nan  # remove invalids

plt.imshow(ndvi, cmap = "RdYlGn")
plt.colorbar(label = "NDVI")
plt.title("NDVI Example")
plt.show()
```
