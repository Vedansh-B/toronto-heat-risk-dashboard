```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import glob

tif_files = glob.glob("../data/raw/modis-terra vegetation indices/nvdi/MOD13Q1.061__250m_16days_NDVI_2024*.tif")

with rasterio.open(tif_files[0]) as src:
    ndvi = src.read(1)  
    profile = src.profile

ndvi = ndvi.astype(float)
ndvi[ndvi < -2000] = np.nan  # remove invalids

plt.imshow(ndvi, cmap = "RdYlGn")
plt.colorbar(label = "NDVI")
plt.title("NDVI Example")
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

shapefile_path = "../data/raw/city boundaries/citygcs_regional_mun_wgs84.shp"

gdf = gpd.read_file(shapefile_path)

print("CRS:", gdf.crs)

fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax = ax, edgecolor = "black", facecolor = "lightgrey")

ax.set_title("City of Toronto", fontsize=14)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

plt.show()
```
